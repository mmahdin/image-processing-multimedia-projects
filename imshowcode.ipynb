```python
# Our Setup, Import Libaries, Create our Imshow Function and Download our Images
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Define our imshow function 
def imshow(title = "Image", image = None, size = 10):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
import cv2
import numpy as np

# Load the image
image = cv2.imread(r'C:\Users\mahdi\Downloads\images\images\bottlecaps.jpg')

# Check if image is loaded successfully
if image is None:
    print("Error: Could not open or find the image.")
    exit()

# Define the size of the larger black background
background_width = image.shape[1] * 2
background_height = image.shape[0] * 2

# Create a black background larger than the original image
def create_black_background(width, height):
    return np.zeros((height, width, 3), dtype=np.uint8)

# Center the image on the black background
def center_image_on_background(image, background_width, background_height):
    h, w = image.shape[:2]
    
    background = create_black_background(background_width, background_height)
    x_offset = (background_width - w) // 2
    y_offset = (background_height - h) // 2
    background[y_offset:y_offset + h, x_offset:x_offset + w] = image
    return background

# Create the larger black background
background = create_black_background(background_width, background_height)

# Center the original image on the larger black background
original_image_on_background = center_image_on_background(image, background_width, background_height)

# Display the original image on the larger black background
imshow('Original Image on Background', original_image_on_background)
```
